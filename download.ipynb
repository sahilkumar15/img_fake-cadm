```python
import os
import gdown
import zipfile
from tqdm import tqdm

file_url = "https://drive.google.com/file/d/1JNMI4RGssgCOl9t05jkUa6imnw5XR5id/view"
zip_path = "checkpoints/checkpoints_pretrain.zip"
extract_dir = "checkpoints"

os.makedirs(extract_dir, exist_ok=True)

if os.path.exists(zip_path):
    os.remove(zip_path)

gdown.download(url=file_url, output=zip_path, quiet=False, fuzzy=True)

print("Downloaded size (MB):", os.path.getsize(zip_path) / (1024 * 1024))

with zipfile.ZipFile(zip_path, "r") as zf:
    members = zf.infolist()
    total_size = sum(m.file_size for m in members)

    with tqdm(total=total_size, unit="B", unit_scale=True, desc="Extracting") as pbar:
        for member in members:
            zf.extract(member, extract_dir)
            pbar.update(member.file_size)

print("Done")
```
